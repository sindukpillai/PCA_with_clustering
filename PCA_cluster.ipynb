```python
import pandas as pd
```

```python
data=pd.read_csv("USArrests.csv")
data.describe()
```

```python
data.columns
```

```python
x=data[[ 'Murder', 'Assault', 'UrbanPop', 'Rape']]
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x.loc[:,:]=sc.fit_transform(x)
```

```python
x
```

```python
from sklearn.decomposition import PCA
pc=PCA()

```

```python
pc_array=pc.fit_transform(x)
```

```python
data_pca=pd.DataFrame(pc_array)
data_pca
```

```python
pc.explained_variance_ratio_*100
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=data_pca[0],y=data_pca[1])
plt.show()
```

```python
plt.scatter(x=data_pca[0],y=data_pca[3])
plt.show()
```

```python
plt.scatter(x=data_pca[1],y=data_pca[3])
plt.show()
```

```python
x=data_pca.loc[:,:]
```

```python

#Kmeans clustering
from sklearn.cluster import KMeans
km=KMeans(n_clusters=5)
```

```python
km.fit(x)
```

```python
cluster_labels=km.predict(x)
cluster_labels
```

```python
import matplotlib.pyplot as plt
plt.scatter(data_pca[0],data_pca[1],c=cluster_labels)
plt.xlabel("murder")
plt.ylabel("Assault")
plt.show()
```

```python
wcss=[]
for i in range(1,11):
  km=KMeans(n_clusters=i)
  km.fit(x)
  inertia=km.inertia_
  wcss.append(inertia)
wcss
```

```python
plt.scatter(x=range(1,11),y=wcss)
plt.xlabel("cluster number")
plt.ylabel("wcss")
plt.show()
```

```python
km=KMeans(n_clusters=5)
km.fit(x)
cluster_lables=km.predict(x)
```

```python
plt.scatter(data_pca[0],data_pca[1],c=cluster_labels)
plt.xlabel("Murder")
plt.ylabel("Assault")
plt.show()
```

```python
#hierarchial
import scipy.cluster.hierarchy as sch
```

```python
dis_matrix=sch.linkage(x,method="average")
```

```python
import matplotlib.pyplot as plt

sch.dendrogram(dis_matrix)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
labels=fcluster(dis_matrix,2.3,criterion="distance")
labels
```

```python
plt.scatter(data_pca[0],data_pca[1],c=labels)
plt.xlabel("murder")
plt.ylabel("assault")
plt.show()
```
